## eda

```python
df['Gender'].value_counts().plot(kind='bar')
```

```python
ax = sns.countplot(x='Gender', data=df)

for i in ax.containers:
    ax.bar_label(i)
```

```python
#females customer are more then males
#lets see whats the amount of sales both have done
```

```python
df.groupby('Gender')['Amount'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
## mostly women have done more shopping & even the amount spend is more then that of men
```

```python
ax = sns.countplot(x='Age Group', data=df, hue='Gender')

for i in ax.containers:
    ax.bar_label(i)
```

```python
## females in every age group have done more shopping then men except in teen-age
## they were sort of similar
```

```python
df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
##mostly 26-35 age group has done shopping & are mostly women
```

```python
df.groupby('State')['Orders'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
df.groupby('State')['Amount'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
##moslty order are from up, maharastra & delhi
```

```python
ax = sns.countplot(x='Marital_Status', data=df)

for i in ax.containers:
    ax.bar_label(i)
```

```python
df.groupby(['Marital_Status','Gender'])['Amount'].sum().plot(kind='bar')
```

```python
## men neither do shopping when married or unmarried
## married women have more purchasing power
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Occupation', data=df)

for i in ax.containers:
    ax.bar_label(i)
```

```python
df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
## It service, healthcare & aviation employees have more purchasing power
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Product_Category', data=df)

for i in ax.containers:
    ax.bar_label(i)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right')
```

```python
sales_by_category_gender = df.groupby(['Product_Category', 'Gender'])['Amount'].sum().reset_index()
ax = sns.barplot(x='Product_Category', y='Amount', hue='Gender', data=sales_by_category_gender)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right')
plt.show()
```

```python
##women are everywhere except power tools ofcourse
##people mostly sepdn on roti, kapd & joota + gadgets also
```
